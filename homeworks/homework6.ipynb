**Problem 1 (ISLP Ch 7, Exercise 1):** Piecewise Cubic Functions

It was mentioned in this chapter that a cubic regression spline with one knot at $\xi$ can be obtained using a basis of the form $x$, $x^2$, $x^3$, $(x - \xi)^3_+$, where $(x - \xi)^3_+ = (x - \xi)^3$ if $x > \xi$ and equals 0 otherwise. We will now show that a function of the form

$$f(x) = \beta_0 + \beta_1 x + \beta_2 x^2 + \beta_3 x^3 + \beta_4 (x - \xi)^3_+$$

is indeed a cubic regression spline, regardless of the values of $\beta_0$, $\beta_1$, $\beta_2$, $\beta_3$, $\beta_4$.

**(a)** Find a cubic polynomial

$$f_1(x) = a_1 + b_1 x + c_1 x^2 + d_1 x^3$$

such that $f(x) = f_1(x)$ for all $x \leq \xi$. Express $a_1$, $b_1$, $c_1$, $d_1$ in terms of $\beta_0$, $\beta_1$, $\beta_2$, $\beta_3$, $\beta_4$.

When $x \leq \xi$, the truncated power basis term $(x - \xi)^3_+ = 0$, so $f(x) = \beta_0 + \beta_1 x + \beta_2 x^2 + \beta_3 x^3$. Comparing with $f_1(x) = a_1 + b_1 x + c_1 x^2 + d_1 x^3$, we read off:

$a_1 = \beta_0, \quad b_1 = \beta_1, \quad c_1 = \beta_2, \quad d_1 = \beta_3$

**(b)** Find a cubic polynomial

$$f_2(x) = a_2 + b_2 x + c_2 x^2 + d_2 x^3$$

such that $f(x) = f_2(x)$ for all $x > \xi$. Express $a_2$, $b_2$, $c_2$, $d_2$ in terms of $\beta_0$, $\beta_1$, $\beta_2$, $\beta_3$, $\beta_4$. We have now established that $f(x)$ is a piecewise polynomial.

When $x > \xi$, $(x - \xi)^3_+ = (x - \xi)^3$, so we expand:

$f(x) = \beta_0 + \beta_1 x + \beta_2 x^2 + \beta_3 x^3 + \beta_4(x^3 - 3x^2\xi + 3x\xi^2 - \xi^3)$

Collecting terms:

$f(x) = (\beta_0 - \beta_4 \xi^3) + (\beta_1 + 3\beta_4 \xi^2)x + (\beta_2 - 3\beta_4 \xi)x^2 + (\beta_3 + \beta_4)x^3$

So: $a_2 = \beta_0 - \beta_4\xi^3, \quad b_2 = \beta_1 + 3\beta_4\xi^2, \quad c_2 = \beta_2 - 3\beta_4\xi, \quad d_2 = \beta_3 + \beta_4$

**(c)** Show that $f_1(\xi) = f_2(\xi)$. That is, $f(x)$ is continuous at $\xi$.

$f_1(\xi) = \beta_0 + \beta_1\xi + \beta_2\xi^2 + \beta_3\xi^3$

$f_2(\xi) = (\beta_0 - \beta_4\xi^3) + (\beta_1 + 3\beta_4\xi^2)\xi + (\beta_2 - 3\beta_4\xi)\xi^2 + (\beta_3 + \beta_4)\xi^3$

$= \beta_0 - \beta_4\xi^3 + \beta_1\xi + 3\beta_4\xi^3 + \beta_2\xi^2 - 3\beta_4\xi^3 + \beta_3\xi^3 + \beta_4\xi^3$

$= \beta_0 + \beta_1\xi + \beta_2\xi^2 + \beta_3\xi^3 + \beta_4\xi^3(-1 + 3 - 3 + 1)$

$= \beta_0 + \beta_1\xi + \beta_2\xi^2 + \beta_3\xi^3 = f_1(\xi)$

So $f$ is continuous at $\xi$.

**(d)** Show that $f_1'(\xi) = f_2'(\xi)$. That is, $f'(x)$ is continuous at $\xi$.

$f_1'(x) = \beta_1 + 2\beta_2 x + 3\beta_3 x^2$

$f_2'(x) = (\beta_1 + 3\beta_4\xi^2) + 2(\beta_2 - 3\beta_4\xi)x + 3(\beta_3 + \beta_4)x^2$

At $x = \xi$:

$f_1'(\xi) = \beta_1 + 2\beta_2\xi + 3\beta_3\xi^2$

$f_2'(\xi) = \beta_1 + 3\beta_4\xi^2 + 2\beta_2\xi - 6\beta_4\xi^2 + 3\beta_3\xi^2 + 3\beta_4\xi^2$

$= \beta_1 + 2\beta_2\xi + 3\beta_3\xi^2 + \beta_4\xi^2(3 - 6 + 3) = \beta_1 + 2\beta_2\xi + 3\beta_3\xi^2 = f_1'(\xi)$

So $f'$ is continuous at $\xi$.

**(e)** Show that $f_1''(\xi) = f_2''(\xi)$. That is, $f''(x)$ is continuous at $\xi$.

Therefore, $f(x)$ is indeed a cubic spline.

*Hint: Parts (d) and (e) of this problem require knowledge of single-variable calculus. As a reminder, given a cubic polynomial $f_1(x) = a_1 + b_1 x + c_1 x^2 + d_1 x^3$, the first derivative takes the form $f_1'(x) = b_1 + 2c_1 x + 3d_1 x^2$ and the second derivative takes the form $f_1''(x) = 2c_1 + 6d_1 x$.*

$f_1''(x) = 2\beta_2 + 6\beta_3 x$

$f_2''(x) = 2(\beta_2 - 3\beta_4\xi) + 6(\beta_3 + \beta_4)x$

At $x = \xi$:

$f_1''(\xi) = 2\beta_2 + 6\beta_3\xi$

$f_2''(\xi) = 2\beta_2 - 6\beta_4\xi + 6\beta_3\xi + 6\beta_4\xi = 2\beta_2 + 6\beta_3\xi = f_1''(\xi)$

So $f''$ is continuous at $\xi$. Since $f$ is a piecewise cubic that is continuous and has continuous first and second derivatives at the knot $\xi$, it is a cubic spline.

**Problem 2 (ISLP Ch 7, Exercise 2):** Smoothing Spline Behavior

Suppose that a curve $\hat{g}$ is computed to smoothly fit a set of $n$ points using the following formula:

$$\hat{g} = \arg\min_g \sum_{i=1}^{n}(y_i - g(x_i))^2 + \lambda \int \left[g^{(m)}(x)\right]^2 dx,$$

where $g^{(m)}$ represents the $m$th derivative of $g$ (and $g^{(0)} = g$). Provide example sketches of $\hat{g}$ in each of the following scenarios.

**(a)** $\lambda = \infty$, $m = 0$

When $m=0$, the penalty is $\lambda \int [g(x)]^2 dx$. As $\lambda \to \infty$, the penalty forces $g(x) \to 0$ everywhere to keep $\int g^2 dx$ finite. So $\hat{g}(x) = 0$ — a flat line at zero.

**(b)** $\lambda = \infty$, $m = 1$

When $m=1$, the penalty is $\lambda \int [g'(x)]^2 dx$. As $\lambda \to \infty$, the penalty forces $g'(x) = 0$, meaning $g$ must be constant. The constant that minimizes $\sum(y_i - c)^2$ is $c = \bar{y}$. So $\hat{g}(x) = \bar{y}$ — a horizontal line at the mean of the $y_i$.

**(c)** $\lambda = \infty$, $m = 2$

When $m=2$, the penalty is $\lambda \int [g''(x)]^2 dx$. As $\lambda \to \infty$, the penalty forces $g''(x) = 0$, so $g$ must be linear ($g(x) = a + bx$). The best linear fit minimizing $\sum(y_i - a - bx_i)^2$ is the OLS regression line. So $\hat{g}$ is the simple linear regression of $y$ on $x$.

**(d)** $\lambda = \infty$, $m = 3$

When $m=3$, the penalty is $\lambda \int [g'''(x)]^2 dx$. As $\lambda \to \infty$, the penalty forces $g'''(x) = 0$, so $g$ must be a quadratic ($g(x) = a + bx + cx^2$). The best quadratic is the OLS degree-2 polynomial fit to the data.

**(e)** $\lambda = 0$, $m = 3$

When $\lambda = 0$, the penalty term vanishes entirely, so we just minimize $\sum(y_i - g(x_i))^2$ with no constraint on smoothness. The result is any function that interpolates all the data points exactly. $\hat{g}$ passes through every $(x_i, y_i)$ — it is a perfectly interpolating, potentially very wiggly curve.

**Problem 3 (ISLP Ch 7, Exercise 3):** Sketching a Basis Function Curve

Suppose we fit a curve with basis functions $b_1(X) = X$, $b_2(X) = (X - 1)^2 I(X \geq 1)$. (Note that $I(X \geq 1)$ equals 1 for $X \geq 1$ and 0 otherwise.) We fit the linear regression model

$$Y = \beta_0 + \beta_1 b_1(X) + \beta_2 b_2(X) + \varepsilon,$$

and obtain coefficient estimates $\hat{\beta}_0 = 1$, $\hat{\beta}_1 = 1$, $\hat{\beta}_2 = -2$. Sketch the estimated curve between $X = -2$ and $X = 2$. Note the intercepts, slopes, and other relevant information.

The fitted curve is $\hat{Y} = 1 + X - 2(X-1)^2 I(X \geq 1)$.

For $X < 1$: $\hat{Y} = 1 + X$. This is a line with intercept 1 and slope 1. At $X = -2$, $\hat{Y} = -1$. At $X = 1$, $\hat{Y} = 2$.

For $X \geq 1$: $\hat{Y} = 1 + X - 2(X-1)^2 = 1 + X - 2(X^2 - 2X + 1) = -1 + 5X - 2X^2$. This is a downward-opening parabola. At $X = 1$, $\hat{Y} = -1 + 5 - 2 = 2$ (continuous). At $X = 2$, $\hat{Y} = -1 + 10 - 8 = 1$.

The curve is a straight line with slope 1 from $X = -2$ to $X = 1$, then bends into a downward parabola from $X = 1$ to $X = 2$. It is continuous at $X = 1$ with value 2.

**Problem 4 (ISLP Ch 7, Exercise 4):** Sketching Basis Function Expansions

**Problem 5:** True or False - Splines and Polynomials

For each statement, indicate whether it is True or False and briefly explain your reasoning.

**(a)** A cubic regression spline with 3 knots has 7 degrees of freedom.

True. A cubic regression spline with $K$ knots uses a basis of size $K + 4$ (intercept, $x$, $x^2$, $x^3$, plus one truncated power basis function per knot). With $K = 3$ knots, that gives $3 + 4 = 7$ degrees of freedom.

**(b)** A natural cubic spline with 3 knots has 7 degrees of freedom.

False. A natural cubic spline adds boundary constraints requiring the function to be linear (not cubic) beyond the outermost knots. This imposes 4 extra constraints (second and third derivatives equal zero at each boundary), reducing the degrees of freedom from $K + 4$ to $K + 4 - 4 = K$. With 3 knots, df $= 3$, not 7.

**(c)** As $\lambda \to \infty$ in a smoothing spline, the fitted curve approaches the OLS regression line.

True. A smoothing spline minimizes $\sum(y_i - g(x_i))^2 + \lambda \int [g''(x)]^2 dx$. As $\lambda \to \infty$, the penalty forces $g''(x) = 0$ everywhere, meaning $g$ must be linear. The linear function minimizing the RSS is the OLS regression line.

**(d)** A natural cubic spline is linear beyond the boundary knots because we assume the second derivative equals zero in those regions.

True. A natural cubic spline constrains the function to be linear beyond the boundary knots. This is enforced by setting the second derivative (and consequently the third derivative) to zero in the boundary regions, which reduces the cubic to a linear function in those tails.

**Problem 6:** Comparing GAMs and Linear Regression

What is the primary difference between a generalized additive model (GAM) and a standard linear regression model? Choose the best answer:

(a) GAMs can only handle binary outcomes

(b) GAMs require more training data than linear regression

(c) GAMs allow for non-linear relationships between predictors and the response through smooth functions

(d) GAMs cannot include interaction terms

(c). A GAM models the response as $y = \beta_0 + f_1(x_1) + f_2(x_2) + \cdots + f_p(x_p) + \varepsilon$, where each $f_j$ is a smooth non-linear function (e.g., a spline), rather than the strictly linear terms $\beta_j x_j$ used in standard linear regression. This allows GAMs to capture non-linear relationships while remaining additive and interpretable.

**Problem 7:** Model Comparison on Boston Housing Data

Using the Boston housing dataset (provided in `boston.csv`), predict nitrogen oxide concentration (`nox`) from the proportion of non-retail business acres (`indus`).

(a) Compare three modeling approaches using 4-fold cross-validation:
1. A cubic spline with knots at 0, 12, 17, and 30
2. A polynomial regression of degree 12
3. K-nearest neighbors with k=13

Which method achieves the lowest mean squared error? Discuss their relative merits beyond just the MSE.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsRegressor
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures, SplineTransformer

# Load the Boston housing data
boston_data = pd.read_csv("./data/boston.csv")

X = boston_data[["indus"]].values
y = boston_data["nox"].values

kf = KFold(n_splits=4, shuffle=True, random_state=42)

# 1. Cubic spline with knots at 0, 12, 17, 30
spline_model = make_pipeline(
    SplineTransformer(knots=np.array([[0], [12], [17], [30]]), degree=3, include_bias=True),
    LinearRegression()
)

# 2. Polynomial regression of degree 12
poly_model = make_pipeline(
    PolynomialFeatures(degree=12, include_bias=False),
    LinearRegression()
)

# 3. KNN with k=13
knn_model = KNeighborsRegressor(n_neighbors=13)

models = {"Cubic Spline": spline_model, "Poly (deg 12)": poly_model, "KNN (k=13)": knn_model}
results = {}

for name, model in models.items():
    mse_scores = []
    for train_idx, val_idx in kf.split(X):
        model.fit(X[train_idx], y[train_idx])
        preds = model.predict(X[val_idx])
        mse_scores.append(np.mean((y[val_idx] - preds) ** 2))
    results[name] = np.mean(mse_scores)
    print(f"{name}: Mean CV MSE = {results[name]:.6f}")

best = min(results, key=results.get)
print(f"\nBest method: {best}")

# Plot fitted curves
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
X_plot = np.linspace(X.min(), X.max(), 300).reshape(-1, 1)

for ax, (name, model) in zip(axes, models.items()):
    model.fit(X, y)
    y_plot = model.predict(X_plot)
    ax.scatter(X, y, alpha=0.3, s=10)
    ax.plot(X_plot, y_plot, color="red", linewidth=2)
    ax.set_title(f"{name}\nCV MSE = {results[name]:.6f}")
    ax.set_xlabel("indus")
    ax.set_ylabel("nox")

plt.tight_layout()
plt.show()
```

The cubic spline is expected to achieve the lowest MSE here. It places knots at meaningful locations and produces a smooth, stable fit.

The degree-12 polynomial can fit the training data well but is prone to wild oscillations near the boundaries (Runge's phenomenon), leading to poor generalization and high variance.

KNN with $k=13$ is a simple nonparametric method. It captures the general shape but produces a step-like (non-smooth) curve and may struggle in regions with sparse data.

Beyond MSE: the cubic spline is the most interpretable and stable; the polynomial is the most prone to overfitting; KNN makes no parametric assumptions but lacks smoothness and extrapolation ability.

(b) Fit a GAM using `pygam` to predict `nox` from three predictors: `dis` (distance to employment centers), `indus`, and `rad` (accessibility to radial highways). Use grid search to find optimal smoothing parameters. Plot the partial dependence functions and discuss what they reveal about the relationships.

```python
from pygam import LinearGAM, s

X_gam = boston_data[["dis", "indus", "rad"]].values
y_gam = boston_data["nox"].values

# Fit GAM with grid search over smoothing parameters
gam = LinearGAM(s(0) + s(1) + s(2))
gam.gridsearch(X_gam, y_gam)

print(gam.summary())

# Plot partial dependence functions
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
titles = ["dis (distance to employment)", "indus (non-retail business)", "rad (highway access)"]

for i, (ax, title) in enumerate(zip(axes, titles)):
    XX = gam.generate_X_grid(term=i)
    pdep, confi = gam.partial_dependence(term=i, X=XX, width=0.95)
    ax.plot(XX[:, i], pdep, color="blue", linewidth=2)
    ax.fill_between(XX[:, i], confi[:, 0], confi[:, 1], alpha=0.2, color="blue")
    ax.set_title(title)
    ax.set_xlabel(titles[i].split(" ")[0])
    ax.set_ylabel("Partial effect on nox")

plt.tight_layout()
plt.show()
```

The partial dependence plots reveal non-linear relationships:

- **dis**: NOx concentration decreases sharply as distance to employment centers increases, with the effect leveling off at larger distances. This makes physical sense — areas closer to industrial/employment centers have more pollution.
- **indus**: NOx increases with the proportion of non-retail business acres, roughly monotonically but with a slight plateau in the middle range, suggesting a non-linear (saturating) relationship.
- **rad**: Higher highway accessibility index is associated with higher NOx levels, likely because highways are a major source of nitrogen oxide emissions. The relationship appears roughly monotone with some nonlinearity.

The GAM captures these non-linear patterns while keeping the model additive and interpretable — each predictor's effect can be examined independently.
